```python
# Импорт необходимых библиотек
import cv2
import numpy as np
from matplotlib import pyplot as plt
from google.colab import files
```

```python
# Функция для обнаружения лиц с использованием классификатора Haar Cascade
def detect_faces(image_path):
    # Загрузка изображения
    image = cv2.imread(image_path)
    # Преобразование изображения в оттенки серого
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Загрузка классификатора Haar Cascade для обнаружения лиц
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

    # Обнаружение лиц на изображении
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    # Нанесение ограничительных прямоугольников вокруг лиц (bounding boxes)
    for (x, y, w, h) in faces:
        cv2.rectangle(image, (x, y), (x+w, y+h), (255, 0, 0), 2)

    # Отображение изображения с ограничительными прямоугольниками
    plt.figure(figsize=(12, 12))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()

# Загрузка изображения в Colab
uploaded = files.upload()

# Получение пути загруженного изображения
image_path = next(iter(uploaded.keys()))

# Обнаружение лиц на загруженном изображении
detect_faces(image_path)
```
